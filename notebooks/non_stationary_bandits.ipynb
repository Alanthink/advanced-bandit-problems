## Exploring bandits with non-stationary reward distributions

Let us build on the multi armed bandit problem by relaxing the assumption that the reward distributions are stationary. Non-stationary reward distributions change over time, and thus our algorithms have to adapt to them. Let us try adding buffers to an $\epsilon$-greedy policy and Thompson Sampling to solve the problem.

```python
# plotting inline
%matplotlib inline

# importing necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sps
from scipy.stats import beta as beta_dist
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from tqdm import tqdm

# turning off automatic plot showing, and setting style
plt.ioff()
plt.style.use('bmh')
```

## Non-stationary bandits

Let us define some test cases where the reward distributions can change and implement them. In this Notebook, we assume that the rewards $r_k$ are binary, and distributed according to a Bernoulli random variable $r_k \sim Bernoulli(\theta_k)$, where $\theta_k$ is our reward parameter, the expected reward at each round.

1. **Reward parameters change according to some distribution every $K$ rounds:** let us define each of our reward parameters $\theta_k$ as a random variable, say, $\theta_k \sim \mathcal{U}(0,1)$, and sample a new reward parameter vector for our bandits $\theta$ every $K$ rounds. The algoritms must adapt to the new conditions, which suddenly change.

2. **Reward parameters change linearly over time:** in this case, we set two reward parameter vectors, $\theta_0$ and $\theta_1$, and linearly change the actual reward vector from $\theta_0$ and $\theta_1$ during $K$ rounds. So, if $n$ is the number of rounds that we have played so far, our reward vector is $\theta = (1 - \frac{n}{K}) \cdot{} \theta_0 + \frac{n}{K} \cdot{} \theta_1$.

##### Let us implement a class in order to play these games:

```python
# now implementing non stationary MABs on top
class NonStationaryMAB:
    
    # initialization
    def __init__(self, bandit_probs_1, K, bandit_probs_dist, mode='dist', bandit_probs_2=None):
        
        # storing parameters
        self.K = K
        self.bandit_probs_dist = bandit_probs_dist
        self.mode = mode
        self.bandit_probs_1 = np.array(bandit_probs_1)
        self.bandit_probs_2 = np.array(bandit_probs_2)
        
        # initializing actual bandit probabilities
        self.bandit_probs = self.bandit_probs_1
        
    # function for drawing arms
    def draw(self, k, n):
        
        # selecting mode
        if self.mode == 'dist':
            
            # need to update reward parameters?
            if ((n % self.K) == 0) & (n != 0):
                self.bandit_probs = self.bandit_probs_dist.rvs(len(self.bandit_probs))
        
        # selecting mode
        elif self.mode == 'linear':
            
            # updating parameters
            if (n <= self.K) & (n >= 0):
                self.bandit_probs = (1 - n/self.K)*self.bandit_probs_1 + n/self.K*self.bandit_probs_2
        
        # if no known mode is selected, raise error    
        else:
            raise ValueError()
        
        # guaranteeing that values are in the 0, 1 range
        self.bandit_probs = np.clip(self.bandit_probs, 0, 1)
        
        # returning a reward for the current arm
        return np.random.binomial(1, self.bandit_probs[k]), np.max(self.bandit_probs) - self.bandit_probs[k]
```

##### Now, let us implement a function to plot the game results: bandit parameters over time, arm selections, and the cumulative regret plot.

```python
# function to plot game results
def plot_game_results(bandit_probs_array, cumulative_regret_array, selection_array):
    
    # defining colors for each bandit
    bandit_colors = ['red', 'green', 'blue', 'purple']
    
    # opening figure
    fig = plt.figure(figsize=(12,6), dpi=120)
    
    # let us draw a grid to position the plots
    ax1 = plt.subplot2grid((4, 2), (0, 0))
    ax2 = plt.subplot2grid((4, 2), (1, 0))
    ax3 = plt.subplot2grid((4, 2), (2, 0))
    ax4 = plt.subplot2grid((4, 2), (3, 0))
    ax5 = plt.subplot2grid((4, 2), (0, 1), rowspan=2)
    ax6 = plt.subplot2grid((4, 2), (2, 1), rowspan=2)
    
    # first plot for first bandit parameter
    ax1.plot(range(bandit_probs_array.shape[0]), bandit_probs_array[:,0], color=bandit_colors[0])
    ax1.fill_between(range(bandit_probs_array.shape[0]), bandit_probs_array[:,0], alpha=0.3, color=bandit_colors[0])
    ax1.set_ylim(-0.05, 1.05) 
    ax1.set_title('Parameter for bandit 0', fontsize=10); ax1.set_ylabel('$\\theta_0$'); 

    # second plot for second bandit parameter
    ax2.plot(range(bandit_probs_array.shape[0]), bandit_probs_array[:,1], color=bandit_colors[1])
    ax2.fill_between(range(bandit_probs_array.shape[0]), bandit_probs_array[:,1], alpha=0.3, color=bandit_colors[1])
    ax2.set_ylim(-0.05, 1.05)
    ax2.set_title('Parameter for bandit 1', fontsize=10); ax2.set_ylabel('$\\theta_1$'); 

    # third plot for third bandit parameter
    ax3.plot(range(bandit_probs_array.shape[0]), bandit_probs_array[:,2], color=bandit_colors[2])
    ax3.fill_between(range(bandit_probs_array.shape[0]), bandit_probs_array[:,2], alpha=0.3, color=bandit_colors[2])
    ax3.set_ylim(-0.05, 1.05)
    ax3.set_title('Parameter for bandit 2', fontsize=10); ax3.set_ylabel('$\\theta_2$'); 

    # fourth plot for fourth bandit parameter
    ax4.plot(range(bandit_probs_array.shape[0]), bandit_probs_array[:,3], color=bandit_colors[3])
    ax4.fill_between(range(bandit_probs_array.shape[0]), bandit_probs_array[:,3], alpha=0.3, color=bandit_colors[3])
    ax4.set_ylim(-0.05, 1.05)
    ax4.set_title('Parameter for bandit 3', fontsize=10); ax4.set_ylabel('$\\theta_3$'); 
    ax4.set_xlabel('Round', fontsize=10)
    
    # now, we plot the cumulative regret for these guys
    ax5.plot(range(cumulative_regret_array.shape[0]), cumulative_regret_array, 'k-')
    ax5.fill_between(range(cumulative_regret_array.shape[0]), cumulative_regret_array, alpha=0.3, color='black')
    ax5.set_title('Cumulative regret for the episode', fontsize=10); ax5.set_xlabel('Round', fontsize=10)
    ax5.set_ylabel('Cumulative regret', fontsize=10)
    
    # finally, we plot the arm choices over time
    #ax6.scatter(range(cumulative_regret_array.shape[0]), selection_array, c=[bandit_colors[i] for i in selection_array], marker='s', s=10)
    a = np.cumsum(selection_array, axis=1)
    ax6.fill_between(range(a.shape[1]), y1=a[3,:], y2=0, color=bandit_colors[3], alpha=0.9)
    ax6.fill_between(range(a.shape[1]), y1=a[2,:] + a[3,:], y2=a[3,:], color=bandit_colors[2], alpha=0.9)
    ax6.fill_between(range(a.shape[1]), y1=a[1,:] + a[3,:] + a[2,:], y2=a[3,:] + a[2,:], color=bandit_colors[1], alpha=0.9)
    ax6.fill_between(range(a.shape[1]), y1=a[0,:] + a[3,:] + a[2,:] + a[1,:], y2=a[3,:] + a[2,:] + a[1,:], color=bandit_colors[0], alpha=0.9)
    ax6.set_title('Arm selection over time', fontsize=10); 
    ax6.set_xlabel('Round', fontsize=10)
    #ax6.set_ylabel('Bandit', fontsize=10); ax6.set_yticks([0,1,2,3])
    ax6.set_ylabel('Number of rounds', fontsize=10)
    
    # setting a tight layout
    fig.tight_layout()
    
    # returning the figure
    return fig
```

## Policies

Let us implement a $\epsilon$-greedy policy along with Thompson Sampling. For both policies, we implement a **buffer**, which is the size of the game history we want the policies to remember to make their decisions. The smaller the buffer is, the algorithm explores more and can better adapt to the changes in the parameters. However, it may also waste useful opportunity to exploit the system. 

There is a trade-off between these two issues in deciding the size of the buffer. For people doing data science in industry, the size of the buffer may be defined by business expert knowledge (the number of days that the market takes to react to a pricing decision, for instance).

```python
# e-greedy policy
class eGreedyPolicy:
    
    # initializing
    def __init__(self, epsilon, buffer_size):
        
        # saving parameters
        self.epsilon = epsilon
        self.buffer_size = buffer_size
    
    # choice of bandit
    def choose_bandit(self, k_array, reward_array, n_bandits):
        
        # number of plays
        n_plays = int(k_array.sum())
        
        # limiting the size of the buffer
        reward_array = reward_array[:,:n_plays][:,-self.buffer_size:]
        k_array = k_array[:,:n_plays][:,-self.buffer_size:]
       
        # sucesses and failures
        success_count = reward_array.sum(axis=1)
        total_count = k_array.sum(axis=1)
         
        # ratio of sucesses vs total
        success_ratio = success_count/total_count
        
        # choosing best greedy action or random depending with epsilon probability
        if np.random.random() < self.epsilon:
            
            # returning random action, excluding best
            return np.random.choice(np.delete(list(range(N_BANDITS)), np.argmax(success_ratio)))
        
        # else return best
        else:
            
            # returning best greedy action
            return np.argmax(success_ratio)    
```

```python
# thompson sampling
class TSPolicy:
    
    # initializing
    def __init__(self, buffer_size):
        
        # nothing to do here
        self.buffer_size = buffer_size
    
    # choice of bandit
    def choose_bandit(self, k_array, reward_array, n_bandits):
        
        # number of plays
        n_plays = int(k_array.sum())
        
        # limiting the size of the buffer
        reward_array = reward_array[:,:n_plays][:,-self.buffer_size:]
        k_array = k_array[:,:n_plays][:,-self.buffer_size:]

        # list of samples, for each bandit
        samples_list = []
        
        # sucesses and failures
        success_count = reward_array.sum(axis=1)
        failure_count = k_array.sum(axis=1) - success_count
                    
        # drawing a sample from each bandit distribution
        samples_list = [np.random.beta(1 + success_count[bandit_id], 1 + failure_count[bandit_id]) for bandit_id in range(n_bandits)]
                                
        # returning bandit with best sample
        return np.argmax(samples_list)    
```

## Evaluation function

Let us build code that allows us to evaluate our policies. In order to evaluate the policies at the same episode (with the same expected rewards), we let the policies play in parallel.

```python
# our game 
def play_nonstat_mab(decision_policy_dict, nsmab, n_draws):
    
    # number of bandits
    n_bandits = len(nsmab.bandit_probs)

    # results dict
    res_dict = {}
    for key in decision_policy_dict.keys():
        res_dict[key] = {'k_array': np.zeros((n_bandits,n_draws)), 
                         'reward_array': np.zeros((n_bandits,n_draws)),
                         'k_list': [], 'reward_list': [], 'bandit_probs_list': [], 'regret_list':[]}

    # colors for each bandit
    bandit_colors = ['red', 'green', 'blue', 'purple']

    # loop generating draws
    for draw_number in range(n_draws):
        
        # loop for each policy tested
        for i, key in enumerate(decision_policy_dict.keys()):
            
            # conditional that stops parameters from updating two times
            if (draw_number % nsmab.K) == 0:
                draw_number = draw_number + i
            
            # choosing arm and drawing
            k = decision_policy_dict[key](res_dict[key]['k_array'], res_dict[key]['reward_array'], n_bandits)
            reward, regret = nsmab.draw(k, draw_number)

            # record information about this draw
            res_dict[key]['k_list'].append(k)
            res_dict[key]['regret_list'].append(regret)
            res_dict[key]['reward_list'].append(reward)
            res_dict[key]['bandit_probs_list'].append([nsmab.bandit_probs[i] for i in range(4)])
            res_dict[key]['k_array'][k, draw_number] = 1
            res_dict[key]['reward_array'][k, draw_number] = reward
            
    # converting to numpy arrays and accumulating regret
    for i, key in enumerate(decision_policy_dict.keys()):
        res_dict[key]['k_list'] = np.array(res_dict[key]['k_list'])
        res_dict[key]['regret_list'] = np.cumsum(np.array(res_dict[key]['regret_list']))
        res_dict[key]['bandit_probs_list'] = np.array(res_dict[key]['bandit_probs_list'])

    # returning relevant elements
    return res_dict
```

## Challenge 1

Let us measure how well our methods fare in the first challenge.

##### Defining parameters for the bandits and configuring policies and buffer sizes

```python
# instatiating the policies
e_greedy_policy = eGreedyPolicy(0.10, 500)
ts_policy = TSPolicy(500)
```

```python
# dict of policies
policy_dict = {'ts': ts_policy.choose_bandit, 
               'e_greedy': e_greedy_policy.choose_bandit}
```

##### Running several simulations for both policies

```python
# dict for our simulations
simulation_dict = {}

# number of simulations
N_SIMULATIONS = 50

# loop for each simulation
for sim in tqdm(range(N_SIMULATIONS)):
    
    # our initial probabilites
    bandit_probs = sps.uniform(0.0, 1.0).rvs(4)
    
    # instance of non stationary mab class
    nsmab = NonStationaryMAB(bandit_probs, 2500, sps.uniform(0.0, 1.0))
    
    # playing the game with both our policies
    res_dict = play_nonstat_mab(policy_dict, nsmab, 10000)
    simulation_dict[sim] = res_dict
```

##### Let us now check results for an individual game from TS.

```python
# let us check game results
plot_game_results(simulation_dict[0]['ts']['bandit_probs_list'], 
                  simulation_dict[0]['ts']['regret_list'], 
                  simulation_dict[0]['ts']['k_array']);
```

##### Also checking results for an individual game from $\epsilon$-greedy.

```python
# let us check game results
plot_game_results(simulation_dict[0]['e_greedy']['bandit_probs_list'], 
                  simulation_dict[0]['e_greedy']['regret_list'], 
                  simulation_dict[0]['e_greedy']['k_array']);
```

##### Let us now compare regrets over many simulations, for both policies.

```python
# closing all past figures
plt.close('all')

# opening figure to plot regret
plt.figure(figsize=(10, 3), dpi=150)

# loop for each decision policy
for policy in ['e_greedy','ts']:
    
    # calculating average regret over all simulations
    avg_regret = pd.DataFrame([simulation_dict[i][policy]['regret_list'] for i in range(N_SIMULATIONS)]).mean(axis=0)
    
    # plotting data
    #plt.plot(np.cumsum(res_dict[policy]['regret_list']), label=policy, linewidth=1.5);
    plt.plot(avg_regret, label=policy, linewidth=1.5);
    
# adding title
plt.title('Comparison of cumulative regret for each method', fontsize=10)

# adding legend
plt.legend(fontsize=8); plt.xticks(fontsize=10); plt.yticks(fontsize=10)

# showing plot
plt.show()
```

The results show that Thompson Sampling compares favourably against the $\epsilon$-greedy policy. The combined visualizations show much better performance of TS in the first game. We can see that TS adapted quickly to changing parameters, maintaining a good dose of exploration. Seems like the $\epsilon$-greedy policy over-explored in the entire game, so tuning the $\epsilon$ parameters could be an improvement to this Notebook. The comparison plot shows great improvement by using TS, with much smaller final cumulative regret. We can also see some bumps in the plot at the times of sudden parameter change, showing how much time the algorithms take to adapt to the new conditions.

In this Notebook, we studied non-stationary bandits, which the distribution of true expected rewards $\theta$ changes over time. We presented two scenarios: (a) $\theta$ changes suddenly and (b) $\theta$ changes slowly and linearly. In both cases, Thompson Sampling compared favourably to an $\epsilon$-greedy policy. Using a buffer was beneficial such that TS explored more and could quickly adapt to changes in the bandits true parameters. In our final experiment, we tested different buffer sizes and observed TS cumulative regret over many simulations. Small buffer sizes perform little exploitation, with cumulative regret being close to linear. On the other hand, large buffer sizes show great regret results only until the first parameter change, when they make TS very slow to adapt to the new conditions. In our setting, we could find a good tradeoff. In business situations, a subject matter expert could advise on buffer sizes, giving insight on how much the system can change over time.
